```python
# Ejercicio 1
def calcular_energia_final(energia_base, factor):
    factor_minimo = 0.1
    energia_base = float(energia_base)
    factor = float(factor)

    if factor < factor_minimo:
        factor = factor_minimo
    energia_final = energia_base * factor
    return energia_final
calcular_energia_final(100,0.01)
```

```python
assert calcular_energia_final(100,0.5) == 50.0
assert calcular_energia_final(100,0.01) == 10.0 #Se aplica el mÃ­nimo 
assert calcular_energia_final(0,0.5) == 0
```

```python
# Ejercicio 4
def leer_potencial(valor):
    try:
        num = int(valor)
    except ValueError:
        return "Error: la potencia debe ser un numero entero"
    return num
    
leer_potencial("abc")
```

```python
#TESTS 
assert leer_potencial("25") == 25
assert leer_potencial("0") == 0
assert leer_potencial("abc") == "Error: la potencia debe ser un numero entero"
```

```python
# Ejercicio 5
def calcular_puntos(velocidad, choques):
    velocidad =int(velocidad)
    choques = int(choques)
    puntos = velocidad * 2 - choques* 3
    
    if puntos < 0:
        return 0
    return puntos


calcular_puntos(10,2)
```

```python
#TESTS
assert calcular_puntos(10,2)== 14
assert calcular_puntos(0,0)== 0
assert calcular_puntos(5,10)== 0
```
