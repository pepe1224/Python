## Ejercicio 1 ‚Äì Clase Camiseta

Crea la clase `Camiseta` con atributos: `talla`, `color`, `precio`. A√±ade un m√©todo `descripcion()` que devuelva un texto con los datos de la camiseta.

```python
# Escribe tu c√≥digo aqu√≠
class Camiseta:
    def __init__(self, color, talla, precio):
        self.talla = talla
        self.color = color
        self.precio = precio

    def descripcion(self):
        return f"Camiseta {self.color}, talla {self.talla}, {self.precio}‚Ç¨"
    
    def te√±ir(self, color_te√±ir):
        self.color = color_te√±ir

    def rebajas(self, porcentaje):
        self.precio = self.precio * ((100 - porcentaje)/100)
    
```

```python
camiseta1 = Camiseta("roja", "M", 15)
assert camiseta1.descripcion() == "Camiseta roja, talla M, 15‚Ç¨"

camiseta2 = Camiseta('verde', 'S', 5)
assert camiseta2.descripcion() == "Camiseta verde, talla S, 5‚Ç¨"

camiseta2.te√±ir('blanco')
assert camiseta2.color == 'blanco'
assert camiseta2.descripcion() == 'Camiseta blanco, talla S, 5‚Ç¨'

camiseta2.rebajas(20)
assert camiseta2.precio == 4
```

## Ejercicio 2 ‚Äì Clase Viaje

Crea la clase `Viaje` con atributos: `destino`, `dias`, `precio`. A√±ade un m√©todo `coste_diario()` que devuelva el precio dividido entre los d√≠as.

```python
# Escribe tu c√≥digo aqu√≠
class Viaje:
    def __init__(self,destino, dias, precio):
        self.destino = destino
        self.dias = dias
        self.precio = precio

    def coste_diario(self):
        return self.precio / self.dias
    
    def lugar_viaje(self):
        return f"{self.destino}"    
    
```

```python
v = Viaje("Roma", 4, 400)
assert v.coste_diario() == 100
```

## Ejercicio 3 ‚Äì Clase Jugador

Crea la clase `Jugador` con atributos: `nombre`, `dorsal`, `goles`. A√±ade un m√©todo `marcar_gol()` que sume 1 gol al atributo `goles`.

```python
# Escribe tu c√≥digo aqu√≠
class Jugador:
    def __init__(self, nombre, dorsal, goles=0):
        self.nombre = nombre
        self.dorsal = dorsal
        self.goles = goles
    
    def marcar_gol(self):
        self.goles += 1
```

```python
j = Jugador("Alex", 9)
j.marcar_gol()
j.marcar_gol()
assert j.goles == 2
```

## Ejercicio 4 ‚Äì Clase Cancion

Crea la clase `Cancion` con atributos: `titulo`, `artista`, `duracion`. A√±ade un m√©todo `es_larga()` que devuelva True si la duraci√≥n es mayor a 5 minutos.

```python
# Escribe tu c√≥digo aqu√≠
class Cancion:
    def __init__(self, titulo, artista, tiempo=5):
        self.titulo = titulo
        self.artista = artista
        self.tiempo = tiempo
    def es_larga():
        return self.tiempo >= 5 
```

```python
c = Cancion("Epic Song", "DJ Max", 6)
assert c.es_larga() is True
```

## Ejercicio 5 ‚Äì Clase Mascota

Crea la clase `Mascota` con atributos: `nombre`, `tipo`, `energia`. A√±ade m√©todos: `jugar()` que reste 10 a energ√≠a y `dormir()` que sume 20 a energ√≠a.

```python
# Escribe tu c√≥digo aqu√≠. La energ√≠a se inicializa en 50
class Mascota:
    def __init__(self, nombre, tipo, energia=50):
        self.nombre = nombre
        self.tipo = tipo
        self.energia = energia

    def jugar(self):
        self.energia -= 10

    def dormir(self):
        self.energia += 20
```

```python
m = Mascota("Luna", "perro")
m.jugar()
m.dormir()
assert m.energia == 60
```
