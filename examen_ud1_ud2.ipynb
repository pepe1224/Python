### 1ï¸âƒ£ Suma de los nÃºmeros del 1 al 50 que sean mÃºltiplos de 5
**Problema:** Escribe una funciÃ³n `suma_multiplos_5()` que calcule la suma de todos los mÃºltiplos de 5 entre 1 y 50 (ambos incluidos).

```python
def suma_multiplos_5():
    contador = 0
    for i in range(1, 51):
        if i % 5 == 0:
            contador += i
    return contador
suma_multiplos_5()
```
