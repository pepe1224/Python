```python
class Coche():
    def __init__(self, marca, modelo, km=0):
        self.marca=marca
        self.modelo=modelo
        self.km=km
    
    def recorrer(self,kms):
        self.km = kms + self.km
        return self.km
    def mostrar(self):
        return f"{"Ma0rca : ", self.marca}, {"Modelo : ", self.modelo}, {"Kilometros : ", self.km}"
    
class CocheElectrico(Coche):
    def __init__(self, marca, modelo, km=0, bateria=100):
        super().__init__(marca, modelo, km)
        self.bateria=bateria

    def cargar(self):
        return self.bateria
```
