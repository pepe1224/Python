## 2. Piedra papel tijera

```python
import random
opciones = ["piedra", "papel", "tijera"]
cpu = random.choice(opciones)
user = input("Elige piedra, papel o tijera: ").lower()
print("CPU eligiÃ³:", cpu)
if user == cpu:
    print("Empate")
elif (user == "piedra" and cpu == "tijera") or (user == "papel" and cpu == "piedra") or (user == "tijera" and cpu == "papel"):
    print("Ganaste")
else:
    print("Perdiste")
```

## 9. Memorama

```python
cartas = ["a", "b", "a", "b"]
mostrar = ["_"]*4
while "_" in mostrar:
    print(mostrar)
    i = int(input("Carta 1: "))
    j = int(input("Carta 2: "))
    if cartas[i] == cartas[j]:
        mostrar[i] = cartas[i]
        mostrar[j] = cartas[j]
        print("Â¡Match!")
    else:
        print("No coincide.")
print("Â¡Juego completado!")
```

## 10. Simulador de puntos

```python
puntos=0
acciones={"1":("Resolver ejercicio",10),"2":("Ayudar",20),"3":("Participar",5)}
while True:
    print("1) Ejercicio | 2) Ayuda | 3) ParticipaciÃ³n | 0) Salir")
    op=input("Elige: ")
    if op=="0": break
    accion,valor=acciones[op]
    puntos+=valor
    print(f"{accion} (+{valor}) Total:{puntos}")
print("PuntuaciÃ³n final:", puntos)
```
