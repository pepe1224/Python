**Problema:** Crea `primera_vocal(palabra)` que devuelva la primera vocal encontrada en la palabra. Si no hay vocales, devuelve `None`.

```python
def primera_vocal(palabra):
    vocal = "aeiou"
    for vocal in palabra:
        if vocal in palabra:
            return vocal
    return None
```

### 6Ô∏è‚É£ Clasificar nota con `match`
**Problema:** Usa `match` para devolver la calificaci√≥n textual de una nota (0‚Äì10):
- 0‚Äì4 ‚Üí 'Insuficiente'
- 5 ‚Üí 'Suficiente'
- 6 ‚Üí 'Bien'
- 7‚Äì8 ‚Üí 'Notable'
- 9‚Äì10 ‚Üí 'Sobresaliente'

```python
def calificacion(nota):
    nota = 10
    match nota:
        case 0,4:
            print("Insuficiente")
        case 5:
            print("Suficiente")
        case 6:
            print("Bien")
        case 7,8:
            print("Notable")
        case 9,10:
            print("Sobresaliente")
        
    return calificacion(nota)
```

```python
assert calificacion(3) == 'Insuficiente'
assert calificacion(8) == 'Notable'
assert calificacion(10) == 'Sobresaliente'
```

### 9Ô∏è‚É£ Calcular el factorial de un n√∫mero
**Problema:** Implementa `factorial(n)` usando un bucle `for`. Si `n` es 0, devuelve 1.
- Factorial 4: `4*3*2*1`
- Factorial 3: `3*2*1`

```python
def factorial(n):
    if n == 0:
        return 1
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado
```

```python
assert factorial(5) == 120
assert factorial(6) == 720
assert factorial(4) == 24
assert factorial(3) == 6
assert factorial(0) == 1
```
